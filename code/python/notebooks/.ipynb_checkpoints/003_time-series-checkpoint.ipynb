```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
```

```python
definition = 'max'
threshold = '0.33'
```

```python
seasons = [
    'Summer',
    'Autumn',
    'Winter',
    'Spring'
]

data = []
data_dir = '/g/data/er8/users/cd3022/solar_drought/coincident_time_series/'
file_name = definition + '_' + threshold

for season in seasons:
    df = pd.read_pickle(data_dir + season + '_' + file_name + '.pkl')
    data.append(df)

DF = pd.concat(data)
```

```python
total_pixels = 2_259_695
regional_pixels = 13_210

DF['national'] = DF['national'] / total_pixels
DF['regional'] = DF['regional'] / regional_pixels
```

```python
DF.index = pd.to_datetime(DF.index, dayfirst=True, format='mixed')
DF = DF.sort_index(ascending=True)
```

```python
# ENSO data

nino_file = '/home/548/cd3022/aus-historical-solar-droughts/data/ENSO/nino_3.4.txt'
nino_data = pd.read_csv(nino_file, header=None)

nino_data.rename(columns={0:'start_date', 1:'end_date', 2:'value'}, inplace=True)

nino_data["start_date"] = pd.to_datetime(nino_data["start_date"], format="%Y%m%d")
nino_data["end_date"] = pd.to_datetime(nino_data["end_date"], format="%Y%m%d")

nino_data.set_index('start_date', inplace=True)


nino_start = []
nino_end = []
nina_start = []
nina_end = []

for i in range(1, len(nino_data.value)-1):
    if (nino_data.value.iloc[i] < 0.8) and (nino_data.value.iloc[i] > -0.8):
        if nino_data.value.iloc[i+1] >= 0.8:
            nino_start.append(nino_data.index[i+1])
        if nino_data.value.iloc[i+1] <= -0.8:
            nina_start.append(nino_data.index[i+1])

    if (nino_data.value.iloc[i] >= 0.8) and (nino_data.value.iloc[i+1] < 0.8):
        nino_end.append(nino_data.index[i])
    if (nino_data.value.iloc[i] <= -0.8) and (nino_data.value.iloc[i+1] > -0.8):
        nina_end.append(nino_data.index[i])
```

```python
# Plot the data
fig, axes = plt.subplots(2, 1, figsize = (12,8))

for i, col in enumerate(['national', 'regional_mean']):
    ax = axes[i]
    ax.plot(DF[col], color='black')
      

    # if col == 'national':
        # ax.set_ylim(0,0.4)
    if col == 'regional':
        ax.set_ylim(0,1)
    if col == 'regional_mean':
        ax.set_ylim(0,8)

    for i in range(len(nino_start)):
        ax.axvspan(nino_start[i], nino_end[i], color='r', alpha=0.3)

    for i in range(len(nina_start)):
        ax.axvspan(nina_start[i], nina_end[i], color='b', alpha=0.3)

    ax.set_xlim(pd.Timestamp('2015-07-07'), pd.Timestamp('2024-09-30'))

# Add labels and show plot
# Set x-ticks at 3-month intervals
# tick_range = pd.date_range(start=DF.index.min(), end=DF.index.max(), freq="3MS")  # 3MS means 3-month start
# ax.set_xticks(tick_range)
# ax.set_xticklabels(tick_range.strftime("%b %Y"), rotation=45)

# Add y label
fig.text(0.04, 0.5, 'Number of Droughts', va='center', rotation='vertical', fontsize=16)  # y-label


plt.tight_layout(rect=[0.05, 0.05, 1, 1])
plt.savefig('/home/548/cd3022/aus-historical-solar-droughts/figs/coincident_time_series/' + file_name + '.png')
plt.show()
```

## Regional time series

```python
seasons = [
    'Summer',
    'Autumn',
    'Winter',
    'Spring'
]

data = []
data_dir = '/g/data/er8/users/cd3022/solar_drought/coincident_time_series/'
reg_file_name = definition + '_' + threshold + '_REGIONS'

for season in seasons:
    df = pd.read_pickle(data_dir + season + '_' + reg_file_name + '.pkl')
    data.append(df)

reg_DF = pd.concat(data)

reg_DF.index = pd.to_datetime(reg_DF.index, dayfirst=True, format='mixed')
reg_DF = reg_DF.sort_index(ascending=True)
```

```python
for col_name in reg_DF.columns:
    if col_name[1] == 'G':
        new_name = col_name[2:]
    else:
        new_name = col_name[1:-1]
    reg_DF = reg_DF.rename(columns={col_name: new_name})
```
