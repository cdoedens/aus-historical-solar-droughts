```python
from compute_solar_single_cell import solar_workflow
import xarray as xr
from datetime import datetime, timedelta

import os
import requests
import pandas as pd

import matplotlib.pyplot as plt
```

# API Data

```python
def get_site_data(site_id, start_date, end_date):
    
    # Request the generation data
    response = requests.get(
        f"https://api.diagno.energy/research/devices/{site_id}/generation",
        headers={"Authorization": "ce586214-116d-4d28-a525-b783908a19ef"},
        params={"startDate": f"{start_date}T00:00:00+11:00", "endDate": f"{end_date}T00:00:00+11:00"}
    )
    data = response.json()

    df = pd.DataFrame(list(data['data'].items()), columns=['timestamp', 'generation'])
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df.set_index('timestamp', inplace=True)
    df.index = df.index.tz_convert('UTC').tz_localize(None)
    # df.index = df.index.tz_localize(None)

    return df
```

```python
df_meta = pd.read_csv('/g/data/er8/users/cd3022/solar_drought/verification/api_solar.csv')
```

```python
df_meta.iloc[0].latitude
```

```python
site_id = df_meta.iloc[0].id

start_date = "2024-06-01"
end_date = "2024-06-08"

df = get_site_data(site_id, start_date, end_date)

lat = df_meta.iloc[0].latitude
lon = df_meta.iloc[0].longitude
```

```python
start_dt = datetime.strptime(start_date, "%Y-%m-%d")
end_dt = datetime.strptime(end_date, "%Y-%m-%d")

# tilt mmust be a list, where first item is "fixed" or "single_axis"
# if "fixed", second and second items must be the tilt angle and then azimuth
tilt = ['fixed', 0, 0]

# Generate a list of dates
date_range = [start_dt + timedelta(days=i) for i in range((end_dt - start_dt).days + 1)]

data = []

# Loop over the dates
for date in date_range: 
    date_s = date.strftime('%Y/%m/%d')
    data.append(solar_workflow(date_s, tilt, lat, lon)) 

ds = xr.concat(data, dim='time')
```

```python
rated_capacity_workflow = 219.656729124
rated_capacity_api = 54480
ds_cap = ds.apply(lambda x: x * rated_capacity_api / rated_capacity_workflow)
```

```python
# Convert df column to xarray and rename dimension to 'time'
df_da = df['generation'].to_xarray().rename({'timestamp': 'time'})

# Interpolate to match ds time coordinate
# ds_cap['actual'] = ds_cap['actual'].interp(time=df_da.time)
ds_cap['generation'] = df_da.interp(time=ds_cap.time)

time_utc = pd.to_datetime(ds_cap.time.values)
time_aest = time_utc.tz_localize("UTC").tz_convert("Australia/Brisbane")
time_aest_naive = time_aest.tz_convert("Australia/Brisbane").tz_localize(None)
ds_cap = ds_cap.assign_coords(time=("time", time_aest_naive))
```

```python
plot_ds = ds_cap.sel(time=slice('2024-06-01', '2024-06-06'))

plt.figure(figsize=(16,5))
plt.plot(plot_ds.time, plot_ds.actual, label='Himawari/PVLib model')
plt.plot(plot_ds.time, plot_ds.generation, label='Generation Data')
plt.xlabel('time')
plt.ylabel('AC Power Output (W)')
plt.legend()
```
