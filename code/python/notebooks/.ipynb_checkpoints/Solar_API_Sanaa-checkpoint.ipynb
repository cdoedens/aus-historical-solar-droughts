# Plausible data description based on common practices in solar energy data collection  
*Note: Please consult the data provider for the definitive definitions.*

### Data provider (main): Clement Meers <clement.meers@diagno.energy>  
### Data provider (contact): Earl Duran <earl.duran@diagno.energy>; Chris Martell <chris.martell@gses.com.au>

<span style="font-size:1.1em">

**id:** unique identifier encoded or anonymised for privacy. It references a specific solar site or customer record.

**tilt:** The tilt angle (in degrees) of the solar panels relative to the horizontal plane. For example, a tilt of 10° means the panels are inclined 10° above horizontal.

**azimuth:** The compass direction the solar panels are facing, typically measured in degrees from true north. An azimuth of 0° often represents due north, 90° represents east, 180° represents south, etc. North-facing panels (0°) are common in the Southern Hemisphere to maximize sun exposure.

**acSizeWatts:** The rated capacity of the system on the AC (alternating current) side, in watts (W). This can sometimes be the inverter’s maximum output rating.

**dcSizeWatts:** The rated capacity of the system on the DC (direct current) side, in watts. Typically represents the total rated wattage of all solar panels before conversion to AC by the inverter.

**lastHeardUTC:** The last time (in UTC) the system reported data or was “heard” from by the monitoring platform. A value of `None` suggests the system has not recently (or ever) sent data or the data is missing.

**latitude:** Geographic latitude (in decimal degrees) of the solar installation.

**longitude:** Geographic longitude (in decimal degrees) of the solar installation.

**nearestBomStation:** The code/name of the nearest Bureau of Meteorology (BoM) station in Australia, used to correlate local weather data (e.g., irradiance, temperature) with system performance.

**loss:** A measure of system losses, expressed as a fraction of the total energy produced. For example, a loss of 0.2 indicates that 20% of the energy is lost due to factors such as panel inefficiency, wiring, temperature, or shading.

</span>

#### From the script below, you can find devices that are near each other, if we’re receiving data and the metadata for the device. If lastHeardUTC is None, we don’t have any data for the device.
The script groups the solar systems by nearest BoM station. 
Station data can be accessed from BoM website > station data : http://www.bom.gov.au/climate/data/#mapoption

```python
import json

# Count the unique nearestBomStation values
unique_stations = df['nearestBomStation'].unique()
print("Number of unique nearestBomStation:", len(unique_stations))

# Create a dataframe that shows for each nearestBomStation the number of devices near that station
station_counts_df = df.groupby('nearestBomStation').agg(num_devices=('id', 'count')).reset_index()
print("Device counts by nearestBomStation:")
print(station_counts_df)

# Create a dictionary where each nearestBomStation maps to a list of device ids
station_devices_dict = df.groupby('nearestBomStation')['id'].apply(list).to_dict()
print("Dictionary of nearestBomStation with list of device ids:")

# Save the dictionary to a JSON file in the specified directory
json_filepath = os.path.join(directory, "BoM_stations.json")
with open(json_filepath, "w") as json_file:
    json.dump(station_devices_dict, json_file, indent=4)
print(f"Dictionary successfully saved to {json_filepath}")

# Save the station_counts_df dataframe to a CSV file in the specified directory
csv_filepath = os.path.join(directory, "BoMStation.csv")
station_counts_df.to_csv(csv_filepath, index=False)
print(f"DataFrame successfully saved to {csv_filepath}")
```

## This endpoint enables you to query device power data. The time range for a single query is limited to one month.

#### There are a few known issues at this stage:
#### The full device list endpoint sometimes times out. It should work on subsequent retries. A timeout for this endpoint will return an ‘internal server error’.
#### Device IDs returned will change with each query. Any ID returned will work to access the device’s data. You might want to save a list of IDs locally (these will work indefinitely) as this is a consequence of the encryption method I’ve employed.
#### Querying a time range where no data is available will return an ‘internal server error’.


```python
# example -- read data for a specific device between two time periods

import pandas as pd
# Request the generation data
response = requests.get(
    "https://api.diagno.energy/research/devices/RwBb-F1FNbd7KTG9Vk7EBJdnk3qr/generation",
    headers={"Authorization": "ce586214-116d-4d28-a525-b783908a19ef"},
    params={"startDate": "2025-01-01T00:00:00+11:00", "endDate": "2025-01-02T00:00:00+11:00"}
)
data = response.json()

# Extract basic info
device_id = data['deviceId']
start_date = data['startDate']  # string e.g. '2025-01-01T00:00:00+11:00'
end_date   = data['endDate']    # string e.g. '2025-01-02T00:00:00+11:00'

# Convert the generation data (a dict) into a DataFrame
# The dict keys are timestamps and values are generation measurements
df = pd.DataFrame(list(data['data'].items()), columns=['timestamp', 'generation'])

# Convert the timestamp column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Create separate columns for the start time components: year, month, day, hour, and minute
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute

# Sort the dataframe by timestamp (if not already sorted)
df.sort_values('timestamp', inplace=True)

# Compute the duration (in minutes) between consecutive timestamps.
# This assumes that each row's generation is accumulated over that many minutes.
df['duration'] = df['timestamp'].diff().dt.total_seconds().div(60)

# For the first row, which has no previous timestamp, fill the duration with the next row's duration (if available)
if len(df) > 1:
    df.loc[df.index[0], 'duration'] = df['duration'].iloc[1]
else:
    df.loc[df.index[0], 'duration'] = 0

# (Optional) Reset index if you want a clean index column:
df.reset_index(drop=True, inplace=True)

# Define the directory where the CSV will be saved
directory = "/g/data/fy29/sho561/Solar/API_data"

# Build a filename from the device id, start date, and end date separated by underscores.
# Since start_date and end_date include characters (e.g., ':', '+', '-') that are not ideal for filenames,
# we remove or replace them.
def clean_date_string(date_str):
    return date_str.replace(":", "").replace("+", "").replace("-", "")

start_clean = clean_date_string(start_date)
end_clean = clean_date_string(end_date)
filename = f"{device_id}_{start_clean}_{end_clean}.csv"

# Build the full file path
filepath = os.path.join(directory, filename)

# Save the DataFrame to CSV without the index column
df.to_csv(filepath, index=False)
print(f"Data saved to {filepath}")
```

```python
import os
import pandas as pd
import requests

# Define directory and CSV path
directory = "/g/data/fy29/sho561/Solar/API_data"
csv_path = os.path.join(directory, "api_solar.csv")

# Read the CSV file
df = pd.read_csv(csv_path)

# Define start_year and construct start and end dates
start_year = 2021
start_date = f"{start_year}-01-01T00:00:00+11:00"
end_date = f"{start_year}-01-02T00:00:00+11:00"

# Initialize list to store results for each device
results = []

# Loop through each row (each device) in the dataframe
for idx, row in df.iterrows():
    device_id = row['id']
    url = f"https://api.diagno.energy/research/devices/{device_id}/generation"
    params = {"startDate": start_date, "endDate": end_date}
    headers = {"Authorization": "ce586214-116d-4d28-a525-b783908a19ef"}
    
    try:
        response = requests.get(url, headers=headers, params=params)
        # If the API request is successful, check if generation data is available
        if response.status_code == 200:
            data = response.json()
            if data.get('data') and len(data.get('data')) > 0:
                # Data is available → mark 1
                results.append({"id": device_id, str(start_year): 1})
            else:
                # No data available → mark 0
                results.append({"id": device_id, str(start_year): 0})
        else:
            results.append({"id": device_id, str(start_year): 0})
    except Exception as e:
        results.append({"id": device_id, str(start_year): 0})

# Create summary DataFrame from results list
summary_df = pd.DataFrame(results)

# Save the summary DataFrame to a CSV file in the specified directory
summary_file = os.path.join(directory, f"summary_generation_{start_year}.csv")
summary_df.to_csv(summary_file, index=False)

print("Summary DataFrame saved to:", summary_file)
```
