# Analysis of 2020

```python
first = '1-8-2015'
# days_per_batch = 30
months_per_batch = 3
num_batches = 20

first_dt = datetime.strptime(first, "%d-%m-%Y")
dates = []
for x in range(num_batches):
    start_dt = first_dt + relativedelta(months = months_per_batch * x)
    start_date = start_dt.strftime("%d-%m-%Y")
    end_dt = start_dt + relativedelta(months = months_per_batch) - relativedelta(days=1)
    end_date = end_dt.strftime("%d-%m-%Y")
    dates.append((start_date, end_date))
    print(start_date, '      ', end_date)
```

```python
data = []
for start, end in dates:
    file_path = f'/g/data/er8/users/cd3022/solar_drought/REZ_tilting/ideal_ratio/NEM_timeseries/{start}___{end}.nc'
    if os.path.exists(file_path):
        data.append(xr.open_dataarray(file_path))
    else:
        print(f"File {file_path} does not exist.")
da = xr.concat(data, dim='time')
```

```python
time_utc = pd.to_datetime(da.time.values)

# Localize to UTC and convert to AEST (handles daylight savings automatically)
time_aest = time_utc.tz_localize("UTC").tz_convert("Australia/Brisbane")

# Assign the converted time back to the xarray DataArray
da_aus_time = da.assign_coords(time=("time", time_aest))
```

```python
seasons = {
    'summer': [12,1,2],
    'autumn': [3,4,5],
    'winter': [6,7,8],
    'spring': [9,10,11]
}

season_colors = {
    "summer": "red",
    "autumn": "orange",
    "winter": "blue",
    "spring": "green"
}
linestyles = ['-', ':', '--', '-.']
```

# Constant below threshold

```python
def constant_below_threshold(da, threshold, linestyle='-'):
    da_droughts = xr.where(da < threshold, 1, 0)
    
    for season in seasons:
        data = da_droughts.where(da_droughts.time.dt.month.isin(seasons[season]), drop=True).data
    
        drought_lengths = [0]
        for i_time in range(1, len(data)):
            drought_past = data[i_time - 1]
            drought_now = data[i_time]
            if (drought_now != 0) and (drought_past != 0):
                drought_now += drought_lengths[i_time - 1]
            drought_lengths.append(drought_now)
    
        length, freq = np.unique(drought_lengths, return_counts = True)
        length = np.array(length[1:]) / 6
        freq = freq[1:]
    
        plt.plot(length, freq, label=season, color=season_colors[season], linestyle=linestyle)
        plt.xlabel('Drought Length (hrs)')
        plt.ylabel('Frequency')
    plt.legend()
    # plt.show()
    # plt.savefig('constant below threshold.png')
```

```python
for i, threshold in enumerate([0.5, 0.4, 0.3, 0.2]):
    plt.title(f'{threshold}')
    plt.xlim(0,6)
    constant_below_threshold(da=da, threshold=threshold, linestyle=linestyles[i])
```

# Mean Below Threshold

```python
def mean_below_threshold(da, threshold):
    drought_dict = {
        'window_size (hrs)':[],
        'summer':[],
        'autumn':[],
        'winter':[],
        'spring':[]
    }
    
    windows = range(1,50)
    
    for window in windows:
    
        drought_dict['window_size (hrs)'].append(window / 6)
        window_mean = da.dropna(dim='time').rolling(time=window, center=True).mean() # dropping nan (overnight) vals has significant effect!
        window_droughts = xr.where(window_mean < threshold, 1, 0)
    
        for season in seasons:
            droughts = window_droughts.where(window_droughts.time.dt.month.isin(seasons[season]), drop=True).data.sum()
            drought_dict[season].append(droughts)
    
    drought_df = pd.DataFrame(drought_dict)
    drought_df.set_index('window_size (hrs)', inplace = True)
    drought_df.plot()
```

```python
for threshold in [0.5, 0.4, 0.3, 0.2]:
    mean_below_threshold(da=da, threshold=threshold)
    plt.title(f'{threshold}')
```

# Annual Variation

```python
for i, year in enumerate([2016, 2017, 2019]):
    da_year = da[da.time.dt.year.isin([year])]
    plt.title('0.4')
    constant_below_threshold(da=da_year, threshold=0.3, linestyle=linestyles[i])
```

# Time of Day

```python
def day_time_heatmap(da):
    
    time = da['time'].to_index()
    
    da.coords['time_of_day'] = ('time', time.time)
    da.coords['day_of_year'] = ('time', time.dayofyear)
    
    da = da.assign_coords(
        time_of_day=("time", time.time),
        day_of_year=("time", time.dayofyear)
    )
    
    df = da.to_dataframe(name="value").reset_index()
    heatmap_df = df.pivot_table(index="day_of_year", columns="time_of_day", values="value", aggfunc="mean")
    
    heatmap_df[heatmap_df == 0] = np.nan
    
    plt.figure(figsize=(15, 8))
    plt.imshow(heatmap_df, aspect="auto", origin="lower", cmap="viridis", vmax=1, vmin=0)
    
    # Set x-axis labels (time of day)
    xticks = np.linspace(0, len(heatmap_df.columns) - 1, 12, dtype=int)  # Select 12 evenly spaced time labels
    xtick_labels = [heatmap_df.columns[i].strftime("%H:%M") for i in xticks]  # Format as HH:MM
    plt.xticks(xticks, xtick_labels, rotation=45)
    
    # Y-axis (Convert Day of Year → Month-Day)
    yticks = np.linspace(0, len(heatmap_df.index) - 1, 10, dtype=int)  # Select 10 evenly spaced day labels
    ytick_labels = [(pd.Timestamp(f"2024-01-01") + pd.Timedelta(days=int(heatmap_df.index[i]) - 1)).strftime("%b %d") for i in yticks]
    plt.yticks(yticks, ytick_labels)
    
    plt.xlabel("Time of Day")
    plt.ylabel("Day of Year")
    plt.title(f"Threshold: {threshold}")
    plt.colorbar(label="Frequency")
    plt.tight_layout()
    plt.show()
```
