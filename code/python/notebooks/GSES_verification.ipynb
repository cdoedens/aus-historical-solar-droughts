```python
def get_site_data(site_id, start_date, end_date):
    
    # Request the generation data
    response = requests.get(
        f"https://api.diagno.energy/research/devices/{site_id}/generation",
        headers={"Authorization": "ce586214-116d-4d28-a525-b783908a19ef"},
        params={"startDate": f"{start_date}T00:00:00+10:00", "endDate": f"{end_date}T00:00:00+10:00"}
    )
    data = response.json()

    df = pd.DataFrame(list(data['data'].items()), columns=['timestamp', 'generation'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)
    df.set_index('timestamp', inplace=True)
    df.index = df.index.tz_convert('UTC').tz_localize(None)
    # df.index = df.index.tz_localize(None)

    return df
```

# Find Good Sites for Verification

```python
def find_good_sites(site):
    site_id = df_meta.iloc[site].id
    
    start_date = "2024-01-01"
    end_date = "2024-01-02"
    
    response = requests.get(
        f"https://api.diagno.energy/research/devices/{site_id}/generation",
        headers={"Authorization": "ce586214-116d-4d28-a525-b783908a19ef"},
        params={"startDate": f"{start_date}T00:00:00+10:00", "endDate": f"{end_date}T00:00:00+10:00"}
        )
    data = response.json()
    if not 'data' in data.keys():
        return False
    good_start = data['data'] != {}
    good_end = df_meta.iloc[site].lastHeardUTC > pd.to_datetime('2024-12-31T23:59:00+10:00')
    
    return good_start and good_end
```

```python
futures = {}
for site in range(df_meta.shape[0]):
    future = client.submit(find_good_sites, site)
    futures[future] = site

results = [False] * df_meta.shape[0]  # Pre-fill results

for future in as_completed(futures):
    site = futures[future]
    try:
        results[site] = future.result()
    except Exception as e:
        print(f"Site {site} failed: {e}")
        results[site] = False
```

```python
good_sites = np.where(pd.Series(results) & ~df_meta.acSizeWatts.isnull())[0]
```

# Get Himawari/PVLib Results

```python
start_dt = datetime.strptime(start_date, "%Y-%m-%d")
end_dt = datetime.strptime(end_date, "%Y-%m-%d")

# tilt mmust be a list, where first item is "fixed" or "single_axis"
# if "fixed", second and second items must be the tilt angle and then azimuth
tilt = ['fixed', tilt, azimuth]

# Generate a list of dates
date_range = [start_dt + timedelta(days=i) for i in range((end_dt - start_dt).days + 1)]

data = []

# Loop over the dates
futures = []
for date in date_range: 
    date_s = date.strftime('%Y/%m/%d')
    f = client.submit(solar_workflow, date_s, tilt, lat, lon)
    futures.append(f) 

results = []
for f in as_completed(futures):
    r = f.result()
    results.append(r)

ds = xr.concat(results, dim='time')
```
