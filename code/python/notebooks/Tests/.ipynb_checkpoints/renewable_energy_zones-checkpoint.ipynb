```python
from datetime import datetime, timedelta
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

import geopandas as gpd
import pyogrio
import rasterio
from rasterio.mask import mask
from pyproj import CRS
from shapely.geometry import Point
```

## Himawari Data

```python
year = 2023
month = 1
day = 1
hour = 23
minute = 30
utc_dt = datetime(year, month, day, hour, minute)
dir_dt = utc_dt + timedelta(hours=5, minutes=30)
num_days = 10
maxv=800

dirin='/g/data/rv74/satellite-products/arc/der/himawari-ahi/solar/p1s/latest/'+f"{dir_dt.year:04}"+'/'+f"{dir_dt.month:02}"+'/'+f"{dir_dt.day:02}"+"/"
# file_count = len([f for f in os.listdir(dirin) if os.path.isfile(os.path.join(dirin, f))])

# Find lat/lon indices to use on all datasets
file_dt = utc_dt.strftime("%Y%m%d%H%M")
filename='IDE00326.'+file_dt+'.nc'
dataset = Dataset(dirin+filename)

latitudes = dataset.variables['latitude'][:]
longitudes = dataset.variables['longitude'][:]
irradiance = np.squeeze(dataset.variables['surface_global_irradiance'][:,:,:])
```

## REZ GIS File

```python
rez_file_name = '/home/548/cd3022/aus-historical-solar-droughts/data/boundary_files/Indicative REZ boundaries 2024 GIS data.kml'
rez_file = gpd.read_file(rez_file_name, engine='pyogrio')
```

### REZs with solar

```python
zones_to_ignore = [
    'Q1',
    'N7',
    'N8',
    'N10',
    'N11',
    'V3',
    'V4',
    'V7',
    'V8',
    'T4',
    'S1',
    'S3',
    'S4',
    'S5',
    'S10'
]
zone_codes = rez_file['Name'].str.split().str[0]
```

```python
# Create grid of points
lon_grid, lat_grid = np.meshgrid(longitudes, latitudes)
points = gpd.GeoDataFrame(geometry=gpd.points_from_xy(lon_grid.ravel(), lat_grid.ravel()), crs=rez_file.crs)

# get solar REZs
polygons = rez_file.loc[~zone_codes.isin(zones_to_ignore)]

# Perform spatial join (efficient way to check if a point is in a polygon)
joined = gpd.sjoin(points, polygons, predicate="within", how="left")

# Remove duplicates based on index instead of geometry
joined = joined[~joined.index.duplicated(keep="first")]

# Create mask (True for points inside a polygon, False otherwise)
mask_flat = joined.index_right.notna().to_numpy()
mask = mask_flat.reshape(lon_grid.shape)

# Apply mask to irradiance data
masked_data = np.where(mask, irradiance, np.nan)
```

```python
mask_file = "/home/548/cd3022/aus-historical-solar-droughts/data/boundary_files/REZ_mask.npz"
np.savez_compressed(mask_file, mask=mask)
```

```python
# Plot or process
# Create a figure with Cartopy's GeoAxes
fig, ax = plt.subplots(figsize=(20, 12), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the data
# Use `extent` if needed to zoom into the relevant area
extent = [longitudes.min(), longitudes.max(), latitudes.min(), latitudes.max()]
img = ax.imshow(masked_data, cmap='viridis', origin='lower', extent=extent, transform=ccrs.PlateCarree())

ax.coastlines()
plt.colorbar(img, ax=ax, label='Data Value')
plt.title('GCCSA')
plt.show()
```

# Find mean irradiance for each zone

```python
solar_rez = rez_file.loc[~zone_codes.isin(zones_to_ignore)]
joined_filtered = joined[joined["index_right"].isin(solar_rez.index)]

points['region'] = joined_filtered['Name']

# Flatten irradiance data
irradiance_flat = irradiance.ravel()

# Convert to DataFrame and drop NaNs
df = gpd.GeoDataFrame({"region": points["region"], "irradiance": irradiance_flat})
df = df.dropna()  # Remove points that are not inside any region

region_means = df.groupby("region")["irradiance"].mean()
mean_total = df["irradiance"].mean()

print(mean_total)
```
